```python
import pandas as pd
import numpy as np
import re
def extract_floor_count(value):
    if pd.isna(value):
        return None
    val = str(value).strip()
    match = re.search(r'out of (\d+)', val, re.IGNORECASE)
    if match:
        return int(match.group(1))
    if val.isdigit():
        return int(val)
    return None
df['Original_floor'] = df['Floor'].apply(extract_floor_count)
```
